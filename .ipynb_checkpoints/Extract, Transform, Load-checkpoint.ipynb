```python
import pandas as pd
import numpy as np

crypto_df = pd.read_csv("https://raw.githubusercontent.com/diljeet1994/Python_Tutorials/master/Projects/Advanced%20ETL/crypto-markets.csv")
```

```python
assetCode = ['BTC', 'ETH', 'XRP', 'LTC']

crypto_df['open'] = crypto_df[['open', 'asset']].apply(lambda x: float(x[0])*0.80 if x[1] in assetCode else np.nan, axis = 1)
crypto_df['high'] = crypto_df[['high', 'asset']].apply(lambda x: float(x[0])*0.80 if x[1] in assetCode else np.nan, axis = 1)
crypto_df['low'] = crypto_df[['low', 'asset']].apply(lambda x: float(x[0])*0.80 if x[1] in assetCode else np.nan, axis = 1)
crypto_df['close'] = crypto_df[['close', 'asset']].apply(lambda x: float(x[0])*0.80 if x[1] in assetCode else np.nan, axis = 1)

crypto_df.dropna(inplace= True)
crypto_df.reset_index(drop= True, inplace= True)
try:
    crypto_df.drop(labels=['slug', 'ranknow', 'volume', 'market', 'close_ratio', 'spread'], inplace = True, axis = 1)
except Exception as error:
    print(error)
finally:
    print('Table has dropped')
crypto_df
```

```python
import sqlite3 as sql

conn = sql.connect('crypto.db')

try:
    conn.execute('DROP TABLE IF EXISTS Crypto')
except Exception as e:
    print(e)
    
try :
    conn.execute("""
        CREATE TABLE Crypto(
        ID INT PRIMARY KEY,
        Asset TEXT NOT NULL,
        Name TEXT NOT NULL,
        Date DATETIME,
        Open FLOAT DEFAULT 0,
        High FLOAT DEFAULT 0,
        Low FLOAT DEFAULT 0,
        Close FLOAT DEFAULT 0
        );""")
except Exception as e:
    print(e)
    print("Table Creation is Failed")
finally:
    conn.close()
```

```python
crypto_list = crypto_df.values.tolist()
conn = sql.connect('crypto.db')
cur = conn.cursor()

try:
    cur.executemany("""
        INSERT INTO Crypto
        (ASSET, NAME, DATE, OPEN, HIGH, LOW, CLOSE)
        VALUES (?, ?, ?, ?, ?, ?, ?)""", crypto_list)
    conn.commit()
    print('Data Inserted Successfully')
except Exception as e:
    print(e)
finally:
    conn.close()
```
