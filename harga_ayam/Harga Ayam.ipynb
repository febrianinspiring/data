```python
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
with pd.ExcelFile('Harga Ayam.xlsx') as xlsx:
    data = pd.read_excel(xlsx, sheet_name = 'Data_cleaned')

#data = data.iloc[36:,:].reset_index()
```

```python
mean_pasar = data['Price (Pasar)'].mean()
median_pasar = data['Price (Pasar)'].median()
mean_hidup = data['Price (Hidup)'].mean()
median_hidup = data['Price (Hidup)'].median()
mean_pasar, median_pasar, mean_hidup, median_hidup
```

```python
# Plot
sns.set(style="whitegrid")
sns.set_color_codes("pastel")

max_value, min_value = data['Datetime'][-1:], data['Datetime'][0]

f, ax = plt.subplots(figsize=(20, 16))

plt.plot(data['Datetime'], data['Price (Pasar)'], color = 'r', linewidth = 3, label = 'Price (Pasar)')
plt.plot(data['Datetime'], data['Price (Hidup)'], color = 'g', linewidth = 3, label = 'Price (Hidup)')
plt.xlim(min_value, max_value)

plt.plot(data['Datetime'], np.full(len(data['Datetime']), mean_pasar, dtype=np.float), label = 'Mean Price (Pasar)', color = 'r', linewidth = 2, linestyle = '-.')
plt.plot(data['Datetime'], np.full(len(data['Datetime']), median_pasar, dtype=np.float), label = 'Median Price (Pasar)', color = 'r', linewidth = 2, linestyle = ':')
plt.plot(data['Datetime'], np.full(len(data['Datetime']), mean_hidup, dtype=np.float), label = 'Mean Price (Hidup)', color = 'g', linewidth = 2, linestyle = '-.')
plt.plot(data['Datetime'], np.full(len(data['Datetime']), median_hidup, dtype=np.float), label = 'Median Price (Hidup)', color = 'g', linewidth = 2, linestyle = ':')

ax.annotate('Mean Price (Pasar)\nRp %.0f' %mean_pasar, xy = (55,mean_pasar), xytext = (65, mean_pasar-2500), size= 15, weight="bold", color='r', arrowprops = dict(facecolor ='r', shrink = 0.05))
ax.annotate('Median Price (Pasar)\nRp %.0f' %median_pasar, xy = (5,median_pasar), xytext = (15, median_pasar+3000), size= 15, weight="bold", color='r', arrowprops = dict(facecolor ='r', shrink = 0.05))
ax.annotate('Mean Price (Hidup)\nRp %.0f' %mean_hidup, xy = (95,mean_hidup), xytext = (100, mean_hidup-5000), size= 15, weight="bold", color='g', arrowprops = dict(facecolor ='g', shrink = 0.05))
ax.annotate('Median Price (Hidup)\nRp %.0f' %median_hidup, xy = (30,median_hidup), xytext = (35, median_hidup+4500), size= 15, weight="bold", color='g', arrowprops = dict(facecolor ='g', shrink = 0.05))

ax.legend(loc = 'upper left', fontsize = 14)
ax.set_xticks(np.arange(0, len(data['Price (Pasar)'])+1, 9))
ax.set_ylabel('Price (Rp)', size = 20)
ax.set_xlabel('Datetime', size = 20)
ax.set_title('Chicken Price in Malang Region (2017-2020)', size = 30)
plt.savefig('Chicken Price in Malang Region (2017 - 2020).png')
```
