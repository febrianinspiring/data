```python
def reg(years, temps):
  x = np.asarray(years).reshape(-1, 1)
  y = np.asarray(temps)
  model = LinearRegression()
  model.fit(x, y)
  intercept = model.intercept_
  slope = model.coef_
  prediction = model.predict(x)
  return prediction, slope, intercept
```

## Temperature Change in Earth Surface

```python
data = pd.read_csv('{/content}/datasets/berkeleyearth/climate-change-earth-surface-temperature-data/GlobalLandTemperaturesByCity.csv')

# Memilih beberapa kolom yang diperlukan
data = data[['dt', 'City', 'AverageTemperature']]

data['dt'] = pd.to_datetime(data['dt'])
data['year'] = data['dt'].map(lambda x: x.year)
data['month'] = data['dt'].map(lambda x: x.month)
data['City'] = data['City']

# Membuat sebuah fungsi yang digunakan untuk menandai musim
def get_season(month):
    if month >= 3 and month <= 5:
        return 'spring'
    elif month >= 6 and month <= 8:
        return 'summer'
    elif month >= 9 and month <= 11:
        return 'autumn'
    else:
        return 'winter'

# Mencari lamanya pengamatan

min_year = 1913
max_year = 2013
years = range(min_year, max_year)

# Menambahkan kolom musim
data['season'] = data['month'].apply(get_season)

spring_temps = []
summer_temps = []
autumn_temps = []
winter_temps = []

for year in years:
    curr_years_data = data[data['year'] == year]
    spring_temps.append(curr_years_data[curr_years_data['season'] == 'spring']['AverageTemperature'].mean())
    summer_temps.append(curr_years_data[curr_years_data['season'] == 'summer']['AverageTemperature'].mean())
    autumn_temps.append(curr_years_data[curr_years_data['season'] == 'autumn']['AverageTemperature'].mean())
    winter_temps.append(curr_years_data[curr_years_data['season'] == 'winter']['AverageTemperature'].mean())

# Membuat Prediksi dengan Regresi
pred_spring, b1_spring, b0_sping = reg(years, spring_temps)
pred_summer, b1_summer, b0_summer = reg(years, summer_temps)
pred_autumn, b1_autumn, b0_autumn = reg(years, autumn_temps)
pred_winter, b1_winter, b0_winter = reg(years, winter_temps)

#Plot
sns.set(style="whitegrid")
sns.set_color_codes("pastel")

f, ax = plt.subplots(figsize=(10, 8))

plt.plot(years, summer_temps, label='Summer', color='r')
plt.plot(years, autumn_temps, label='Autumn', color='orange')
plt.plot(years, spring_temps, label='Spring', color='g')
plt.plot(years, winter_temps, label='Winter', color='b')
plt.plot(years, pred_summer, color='r', linestyle='--')
plt.plot(years, pred_autumn, color='orange', linestyle='--')
plt.plot(years, pred_spring, color='g', linestyle='--')
plt.plot(years, pred_winter, color='b', linestyle='--')

plt.text(0.65,0.77, 'Summer', transform= f.transFigure, size= 12, weight="bold", color='r')
plt.text(0.15,0.60, 'Autumn', transform= f.transFigure, size= 12, weight="bold", color='orange')
plt.text(0.65,0.51, 'Spring', transform= f.transFigure, size= 12, weight="bold", color='g')
plt.text(0.15,0.27, 'Winter', transform= f.transFigure, size= 12, weight="bold", color='b')
plt.text(0.65,0.75, 'Increase in %.3f$^\circ$C / year' %b1_spring, transform= f.transFigure, size= 12, color='r')
plt.text(0.15,0.58, 'Increase in %.3f$^\circ$C / year' %b1_autumn, transform= f.transFigure, size= 12, color='orange')
plt.text(0.65,0.49, 'Increase in %.3f$^\circ$C / year' %b1_spring, transform= f.transFigure, size= 12, color='g')
plt.text(0.15,0.25, 'Increase in %.3f$^\circ$C / year' %b1_winter, transform= f.transFigure, size= 12, color='b')
plt.text(0.70,0.14, 'By :Muhammad Febriantoro', transform= f.transFigure, size= 10, color='gray')
plt.xlim(min_year, max_year)

ax.set_ylabel('Temperature ($^\circ$C)')
ax.set_xlabel('Years')
ax.set_title('Global Temperature Visualization in 100 years (1913 - 2013)')
ax.grid(alpha = 0.3)
#legend = plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), frameon=True, borderpad=1, borderaxespad=1)
```

## Perubahan Suhu di Indonesia

```python
data = pd.read_csv('{/content}/datasets/berkeleyearth/climate-change-earth-surface-temperature-data/GlobalLandTemperaturesByCity.csv')

# Memilih beberapa kolom yang diperlukan
data = data[['dt', 'City', 'Country', 'AverageTemperature']]

data['dt'] = pd.to_datetime(data['dt'])
data['year'] = data['dt'].map(lambda x: x.year)
data['month'] = data['dt'].map(lambda x: x.month)
data['City'] = data['City']

data = data.loc[data['Country'] == 'Indonesia']

# Membuat sebuah fungsi yang digunakan untuk menandai musim
def get_season(month):
    if month >= 4 and month <= 9:
        return 'dry'
    else:
        return 'wet'

# Mencari lamanya pengamatan

min_year = 1913
max_year = 2013
years = range(min_year, max_year)

# Menambahkan kolom musim
data['season'] = data['month'].apply(get_season)

wet_temps = []
dry_temps = []

for year in years:
    curr_years_data = data[data['year'] == year]
    wet_temps.append(curr_years_data[curr_years_data['season'] == 'wet']['AverageTemperature'].mean())
    dry_temps.append(curr_years_data[curr_years_data['season'] == 'dry']['AverageTemperature'].mean())

# Membuat Prediksi dengan Regresi
pred_dry, b1_dry, b0_dry = reg(years, dry_temps)
pred_wet, b1_wet, b0_wet = reg(years, wet_temps)

#Plot
sns.set(style="whitegrid")
sns.set_color_codes("pastel")

f, ax = plt.subplots(figsize=(10, 8))

plt.plot(years, dry_temps, label='Rata-rata suhu di musim kering', color='orange')
plt.plot(years, wet_temps, label='Rata-rata suhu di musim hujan', color='b')
plt.plot(years, pred_dry, color='orange', linestyle='--')
plt.plot(years, pred_wet, color='b', linestyle='--')

plt.xlim(min_year, max_year)

ax.set_ylabel('Suhu rata-rata')
ax.set_xlabel('Tahun')
ax.set_title('Suhu rata-rata Indonesia dari tahun 1913 - 2013')
legend = plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), frameon=True, borderpad=1, borderaxespad=1)
```

## Perubahan Suhu di Kota Malang

```python
data = pd.read_csv('{/content}/datasets/berkeleyearth/climate-change-earth-surface-temperature-data/GlobalLandTemperaturesByCity.csv')

# Memilih beberapa kolom yang diperlukan
data = data[['dt', 'City', 'Country', 'AverageTemperature']]

data['dt'] = pd.to_datetime(data['dt'])
data['year'] = data['dt'].map(lambda x: x.year)
data['month'] = data['dt'].map(lambda x: x.month)
data['City'] = data['City']

data = data.loc[data['Country'] == 'Indonesia']
data = data.loc[data['City'] == 'Malang']

# Membuat sebuah fungsi yang digunakan untuk menandai musim
def get_season(month):
    if month >= 4 and month <= 9:
        return 'dry'
    else:
        return 'wet'

# Mencari lamanya pengamatan

min_year = 1913
max_year = 2013
years = range(min_year, max_year)

# Menambahkan kolom musim
data['season'] = data['month'].apply(get_season)

wet_temps = []
dry_temps = []

for year in years:
    curr_years_data = data[data['year'] == year]
    wet_temps.append(curr_years_data[curr_years_data['season'] == 'wet']['AverageTemperature'].mean())
    dry_temps.append(curr_years_data[curr_years_data['season'] == 'dry']['AverageTemperature'].mean())

# Membuat Prediksi dengan Regresi
pred_dry_malang, b1_dry_malang, b0_dry_malang = reg(years, dry_temps)
pred_wet_malang, b1_wet_malang, b0_wet_malang = reg(years, wet_temps)

#Plot
sns.set(style="whitegrid")
sns.set_color_codes("pastel")

f, ax = plt.subplots(figsize=(10, 8))

plt.plot(years, dry_temps, label='Rata-rata suhu musim kering', color='orange')
plt.plot(years, wet_temps, label='Rata-rata suhu musim hujan', color='b')
plt.plot(years, pred_dry_malang, color='orange', linestyle='--')
plt.plot(years, pred_wet_malang, color='b', linestyle='--')

plt.xlim(min_year, max_year)

ax.set_ylabel('Suhu rata-rata')
ax.set_xlabel('Tahun')
ax.set_title('Suhu rata-rata Kota Malang dari tahun 1913 - 2013')
legend = plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), frameon=True, borderpad=1, borderaxespad=1)
```

```python
cities = data['City'].unique()
ind_cities = {'Banda Aceh', 'Medan', 'Padang', 'Palembang',
              'Jakarta', 'Bandung', 'Semarang', 'Yogyakarta', 'Surabaya', 'Malang',
              'Denpasar', 'Pontianak', 'Makasar','Manado', 'Kupang', 'Ambon'}
```

```python
min_max_majorcity = pd.DataFrame(data)

max_majorcity = min_max_majorcity.groupby(['City']).max()['AverageTemperature']
max_majorcity = max_majorcity.to_frame().reset_index()
max_majorcity.columns = ['City','City Maximum']
max_majorcity = max_majorcity.loc[max_majorcity['City'].isin(ind_cities)]  

max_majorcity
```

```python
f, ax = plt.subplots(figsize=(15, 8))
sns.barplot(x='City',y='City Maximum', data=max_majorcity.reset_index())
texts = ax.set(ylabel="MAX Temperature", xlabel="City", title="Cities maximum temperatures")
```

```python
min_majorcity = min_max_majorcity.groupby(['City'],as_index=True).min()['AverageTemperature']
min_majorcity = min_majorcity.to_frame().reset_index()
min_majorcity.columns = ['City','City Minimum']
min_majorcity = min_majorcity.loc[min_majorcity['City'].isin(ind_cities)]
min_majorcity
```

```python
f, ax = plt.subplots(figsize=(15, 8))
sns.barplot(x='City',y='City Minimum', data=min_majorcity.reset_index())
texts = ax.set(ylabel="MIN Temperature", xlabel="City", title="Cities minimum temperatures")
```

```python
##Merging the two data sets
min_max_majorcity_merge = pd.merge(min_majorcity , max_majorcity)
min_max_majorcity_merge = min_max_majorcity_merge.set_index('City')
##print(min_max_majorcity_merge)

ax = min_max_majorcity_merge.plot(kind='bar',stacked=False, figsize=(15, 8))

texts = ax.set(ylabel="City Max-Min Temperature", xlabel="City", title="Cities Min MAX temperatures")
```
